```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib
%matplotlib inline

file_name = "pisa2012.csv"
columns = ['CNT', 'PV1MATH', 'ST04Q01', 'ANCCLSMAN','MMINS','ST55Q02','ANCMATWKETH','DISCLIMA','WEALTH']
df_raw = pd.read_csv(file_name, sep=',',encoding = 'iso-8859-1', usecols=columns)
```

```python
no_significant_extra_math=['2 or more but less than 4 hours a week',
                           '4 or more but less than 6 hours a week',
                           '6 or more hours a week']

df_raw['Extra math classes'] = np.where(df_raw['ST55Q02']==no_significant_extra_math[0],1.0,0.0)
df_raw['Extra math classes'] = np.where(df_raw['ST55Q02']==no_significant_extra_math[1],1.0,df_raw['Extra math classes'])
df_raw['Extra math classes'] = np.where(df_raw['ST55Q02']==no_significant_extra_math[2],1.0,df_raw['Extra math classes'])

df_raw.head()
```

```python

df=df_raw

# countries=['Denmark', 
#            'Sweden', 
#            'Norway', 
#            'New Zealand', 
#            'Switzerland', 
#            'Australia', 
#            'Canada', 
#            'Iceland', 
#            'Finland']

# df=df_raw[df_raw['CNT'].isin(countries)].dropna(axis=0)
```

```python
# file_name_save="Pisa2012_filtered.csv"
# df.to_csv(file_name_save, sep=',')
```

```python
df_country = df.groupby(['CNT'],as_index=False)['PV1MATH', 'MMINS','ANCCLSMAN','Extra math classes','ANCMATWKETH','DISCLIMA', 'WEALTH'].mean()
df_country = df_country.rename(columns={'CNT':'Country',
                                                    'PV1MATH':'Mean math test score', 
                                                    'ANCCLSMAN':'Mean math teacher management',
                                                    'DISCLIMA': 'Mean classroom Discipline',
                                                    'ANCMATWKETH':'Mean math work ethic',
                                                    'WEALTH':"Mean household wealth",
                                                    "MMINS":"Mean time spent on math (minutes)"})
```

```python
min_value = df_country['Mean household wealth'].min()
max_value = df_country['Mean household wealth'].max()

df_country_scaled=df_country.copy()
df_country_scaled['Mean household wealth']=(df_country['Mean household wealth']-min_value) / (max_value - min_value)

file_name_save="Pisa2012_country_mean_scaled.csv"
df_country_scaled.to_csv(file_name_save, sep=',', index=False)
df_country_scaled
```
